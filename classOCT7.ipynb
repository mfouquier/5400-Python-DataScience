```python
#2 rows 1 column with index 1
ax1 = plt.subplot(2, 1, 1)
ax1.plot(a1, np.sin(a1))
#2 rows 1 column with index 2
ax2 = plt.subplot(2, 1, 2)
ax2.plot(a1, np.cos(a1))
```

```python
#2x2 figure
ax11 = plt.subplot(2,2,1)
ax11.plot(a1,np.sin(a1))
ax12 = plt.subplot(2,2,2)
ax12.plot(a1,np.cos(a1))
ax21 = plt.subplot(2,2,3)
ax21.boxplot(a2)
ax22 = plt.subplot(2,2,4)
ax22.scatter(a2, a2/2)
```

```python
#single figure and axes
fig, ax = plt.subplots()
ax.plot(a1,a1**4)
ax.set_title('Radians Raised to the 4th Power')
```

```python
#single figure 2 axes vertically stacked
fig, ax = plt.subplots(2)
fig.suptitle('Vertically Stacked Radian Equation')
ax[0].plot(a1, a1**2)
ax[0].set_title('Squared')
ax[1].plot(a1, a1**3)
ax[1].set_title('Cubed')
```

```python
fig3, ax = plt.subplots(2,2)
ax[0,0].plot(a1,np.sin(a1), color='y')
ax[0,0].set_title('sin')
ax[1,0].plot(a1,np.cos(a1), color='g')
ax[1,0].set_title('cos')
ax[0,1].plot(a1,np.sin(a1*2), color='r')
ax[0,1].set_title('sin 2x')
ax[1,1].plot(a1,np.cos(a1*2), color='b')
ax[1,1].set_title('cos 2x')
```

```python
fig4 = plt.figure(figsize = [9.0,6.0])
ax = fig4.subplots(2,2)
ax[0,0].plot(a1,np.sin(a1), color='y')
ax[0,0].set_title('sin')
ax[1,0].plot(a1,np.cos(a1), color='g')
ax[1,0].set_title('cos')
ax[0,1].plot(a1,np.sin(a1*2), color='r')
ax[0,1].set_title('sin 2x')
ax[1,1].plot(a1,np.cos(a1*2), color='b')
ax[1,1].set_title('cos 2x')
```

```python
#simple line plot
#equation of a line y = mx + b

fig = plt.figure()
```

```python
x1 = np.arange(0,50)
x1
```

```python
#slope
m1 = .5
#y intercept
b1 = 1
print(m1,b1)
```

```python
y1 = m1*x1 + b1
y1
```

```python
ax = plt.axes()
ax.plot(x1, y1)
```

```python
m2 = .55
b2 = 1.0
y2 = m2*x1 + b2

fig = plt.figure()
ax = plt.axes()
ax.plot(x1, y1)
ax.plot(x1, y2)
```

```python
plt.close()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x1, y1, marker='.', color='g')
ax.plot(x1, y2, marker='x', color='r')
```

```python
plt.close()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x1, y1, marker='.', color='g')
ax.plot(x1, y2, marker='x', color='r')
ax.set_xlim([-5,55])
ax.set_ylim([-5,35])
```

```python
plt.close()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x1, y1, marker='.', color='g', label='X')
ax.plot(x1, y2, marker='x', color='r', label='Y')
ax.set_xlabel('X')
ax.set_ylabel('Y = mX + b')
ax.set_title('Simple Line Example')
ax.legend()
ax.set_xlim([-5,55])
ax.set_ylim([-5,35])
```

```python
plt.close()
```

```python
crng = np.random.RandomState(0)
rng = np.random.default_rng(seed=1)
noise1 = rng.standard_normal(size=50)
noise2 = rng.standard_normal(size=50)
#x1 is a random array a few lines up
x1n = x1 + noise1
x2n = x1 + noise2
b1n = rng.standard_normal(size=50)
b2n = rng.standard_normal(size=50)
y1n = m1*x1n + b1n
y2n = m1*x2n + b2n

colors = crng.rand(50)
sizes = 1000 * crng.rand(50)

fig = plt.figure()
ax = plt.axes()
ax.scatter(x1n, y1n, c=colors, alpha=0.3, s=sizes, cmap='viridis')
ax.scatter(x2n, y2n, marker='x', color='r')
ax.set_xlabel('X')
ax.set_ylabel('Y = mx + b')
ax.set_title('Scatter Plot Example')

```

```python
fileName = 'US_Meat_Production.csv'
df1 = pd.read_csv(fileName, index_col=0, header=0)
```

```python
df1.index
```

```python
df1.columns
```

```python
df1
```

```python
df1.loc['Total'] = df1.sum(axis=0)
```

```python
df1
```

```python
fig = plt.figure()
ax = plt.axes()
year = df1.columns
ax.bar(year, df1.loc['Total'], color='r')
ax.set_ylabel("Meat Production in Metric Tons")
ax.set_xlabel('Year')
ax.set_title('Total Meat Production in Metric Tons')
```
