```python
sns.boxplot(x='Defaulted?', y='Annual Salary', hue='Employed', data=df)
```

```python
#Added Hue to show Employed
sns.boxplot(x='Defaulted?', y='Bank Balance', hue='Employed', data=df)
```

```python
#Removed outliers with showfliers
sns.boxplot(x='Defaulted?', y='Bank Balance', hue='Employed', showfliers=False, data=df)
```

```python
#Add width to make plot wider
sns.boxplot(x='Defaulted?', y='Bank Balance', hue='Employed', showfliers=False, width=.25, data=df)
```

```python
#Added dodge and it overlapped the boxplots
sns.boxplot(x='Defaulted?', y='Bank Balance', hue='Employed', showfliers=False, width=.25, dodge=False, data=df)
```

```python
column_list = ['Bank Balance', 'Annual Salary']
for col in column_list : 
    sns.kdeplot(df[col], shade=True)
```
