```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read in the file, csv, and create a dataframe
df = pd.read_csv('EnglishPremierLeagueDataset-1.csv', header=0)
```

```python
#Output the column names
df.columns
```

```python
#Create a plot with the top 10 scorers, use a bar plot and add title and labels as appropriate: 
#hint since the names are long, rotate them using ax.set_xticklabels(df['Name'], rotation=45, ha='right'). 
#You may get a warning regarding FixFormatter. 
#To eliminate the warning, set the tick first before setting the xtickLabels;
# 1. ax.set_xticks(df[‘some column’])
# 2. ax.set_xticklabels(df[‘some column’], rotation=45, ha='right')
df1 = df.sort_values(by='Goals', ascending=False)
df2 = df1.iloc[:10]
fig = plt.figure()
ax = plt.axes()
ax.bar(df2['Name'], df2['Goals'])
ax.set_title('Top Goal Scorers')
ax.set_xticks(df2['Name'])
ax.set_xticklabels(df2['Name'], rotation=45, ha='right')
ax.set_ylabel('Goals')
```

```python
#Create a plot with the top 10 assist makers, use a bar plot and add title and labels as appropriate: 
#hint since the names are long, rotate them as discussed above in item 3
df3 = df.sort_values(by='Assists', ascending=False)
df4 = df3.iloc[:10]
fig = plt.figure()
ax = plt.axes()
ax.bar(df4['Name'], df4['Assists'])
ax.set_title('Top Assists Makers')
ax.set_xticks(df4['Name'])
ax.set_xticklabels(df4['Name'], rotation=45, ha='right')
ax.set_ylabel('Assists')
```

```python
#Create a box plot with Age; are there any outliers?
fig = plt.figure()
ax = plt.axes()
ax.boxplot(df['Age'])
ax.set_title('Age Dispersion')
#There don't seem to be any outliers. 
```

```python
#Create a histogram showing the distribution of the players’ age
fig = plt.figure()
ax = plt.axes()
ax.hist(df['Age'])
ax.set_title('Age')
ax.set_xlabel('Age')

#Question for class -> What is the default Y axis and how can you make it blank for this graph???
```

```python
#Create a plot with the Clubs and the Total number of Goals scored by each club in descending order, use a bar plot and 
#add title and labels as appropriate: hint since the club names are long, rotate them as discussed above in item 3; 
#also you may want to use the groupby as shown in class
df5 = df['Goals'].groupby(df['Club']).sum().sort_values(ascending=False)
df6 = pd.DataFrame(df5)
fig = plt.figure(figsize=[15,8])
ax = plt.axes()
ax.bar(df6.index, df6['Goals'])
ax.set_title('Total Goals by Club')
ax.set_xticks(df6.index)
ax.set_xticklabels(df6.index, rotation=45, ha='right')
ax.set_ylabel('Total Goals')
```
