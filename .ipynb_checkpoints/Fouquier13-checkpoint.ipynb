```python
# Examine the pair plots for finding possible correlations between features and target.
sns.pairplot(df, hue='Exited', vars=['CreditScore', 'Gender', 'Age', 'Tenure', 'TotalBalance',
       'NumOfProducts', 'HasCreditCard', 'IsActiveCustomer', 'Salary'],palette='mako')
```

```python
# Create a correlation matrix with heatmap as demonstrated in class.
corr = df.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
