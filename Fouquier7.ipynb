```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#setting the style
plt.style.use('ggplot')
```

```python
#Read in the file, csv, and create a dataframe
fileName = 'US_Meat_Production.csv'
df1 = pd.read_csv(fileName, index_col=0, header=0)
```

```python
#Output the index values of the dataframe
df1.index
```

```python
#Output the column names of the dataframe
df1.columns
```

```python
#Append a column called ‘Total’ which contains the sum of the meat production for each year
df1['Total'] = df1.sum(axis=1)
df1['Total'] = df1['Total'].astype('int64')
df1
```

```python
#Plot the Total (y axis) against the Year (x axis); hint – use the dataframe’s index for the x axis, 
#and dataframe’s ‘Total’ column for the y axis; Use 'Year' for the X label, 
#‘Meat Production in Tonnes' for the Y label and 'Total Meat Production per Year' for the title

fig = plt.figure()
ax = plt.axes()
year = df1.index
ax.plot(year, df1['Total'], color='r')
ax.set_ylabel("Meat Production in Metric Tonnes")
ax.set_xlabel('Year')
ax.set_title('Total Meat Production per Year')

```

```python
#Create a 3 row, 2 column subplots using the first 6 columns in the dataframe 
#i.e. ignore ‘Horse’; Use 'Year' for the X label; ‘MT’ for the Y label and animal type for the title; 
#Make sure that you size the figure to fit all of the subplots, labels, and titles; Use a different color for each line.

fig4 = plt.figure(figsize = [15.0,12.0])
ax = fig4.subplots(3,2)
year = df1.index
ax[0,0].plot(year,df1['Sheep and Goat'], color='y')
ax[0,0].set_title('Sheep and Goat')
ax[1,0].plot(year,df1['Beef and Buffalo'], color='g')
ax[1,0].set_title('Beef and Buffalo')
ax[0,1].plot(year,df1['Pig'], color='r')
ax[0,1].set_title('Pig')
ax[1,1].plot(year,df1['Game'], color='b')
ax[1,1].set_title('Game')
ax[2,0].plot(year,df1['Duck'], color='c')
ax[2,0].set_title('Duck')
ax[2,1].plot(year,df1['Poultry'], color='m')
ax[2,1].set_title('Poultry')
```
