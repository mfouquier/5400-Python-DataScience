```python
#Step 3 - Data Preparation
#Suppose the data is contained in a vector, how do you convert to a matrix?
```

```python
#Let's generate some data which we will use later for the model
#define the seed for the random number generator
#RandomState also exposes a number number of methods for generating random numbers 
#drawn from a variety of probability distributions.
rng = np.random.RandomState(1)
```

```python
#Using rand generates numbers from the uniform distribution, [0,1]
x = rng.rand(50)
x
```

```python
x.shape
```

```python
#shape (50,) means the array is indexed by a single index which runs from 0 to 49
#it is not a 2 dimensional table which is required by scikit-learn
#let's fix
#we can use the keyword 'newaxis'  to increase the dimensionality of an array
X = x[:,np.newaxis]
```

```python
X[:10]
```

```python
X.shape
```

```python
W = x[np.newaxis,:]
W
```

```python
W.shape
```

```python
#Let's scale X to be numbers between 0 and 10
X = X * 10
```

```python
X[:10]
```

```python
#Recall the equation of a line
# y = mX + b
# where m is the slope of the line and b is the y intercept
#Let's generate y and add some noise to b
#Let's let m = 2 and b = 1
# Let's use the normal random number generator to add noise to b
m = 2
b = 1
```

```python
b = 1 + rng.randn(50)
b
```

```python
b = b[:,np.newaxis]
b[:10]
```

```python
y = m * X + b
```

```python
plt.scatter(X,y)
```

```python
#Linear Regression models are a good starting point for
#regression tasks.
#Simplest form of linear regression model is fitting a
#straight line to the data
#Now let's fit a straight line to the data by
#using scikit-learn's Linear Regression estimator to fit
#this data and construct the best fit line
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Let's create a model
model = LinearRegression(fit_intercept = True)
```

```python
model
```

```python
model.fit(X,y)
```

```python
#Let's use to predict and create the line using a 
#number of x data points
```

```python
xfit = np.linspace(0,10,1000)
```

```python
xfit[:10]
```

```python
#The xfit data is not in the correct format so remember to correct
#when we call the model
#Let's predict y values
yfit = model.predict(xfit[:,np.newaxis])
```

```python
plt.scatter(X,y)
plt.plot(xfit,yfit,c='r')
```

```python
#The slope and the intercept of the data are contained in the
#model's fit attributes, which in scikit-learn are always marked by 
#a trailing underscore
#These values are close to what we specified for the slope and y-intercept
#when we created the data
```

```python
model.coef_
```

```python
model.intercept_
```

```python
#get the coefficient of determination of the prediction. 
#X is your test data and y is the true value
#Best possible score is 1.0
#https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#
score = model.score(X[40:],y[40:])
score
```

```python
#In addition to fitting a straight line to the data, the LinearRegression 
#estimator can handle multidimensional linear models of the form:
#y = w1X1 + w2X2 +... + w0
#where there are multiple x values or features
#with 2 features, the model fits a plane in 3 dimensions
#with more than 2 features, the models fits a hyper-plane in 
#higher dimensions
#Why did I switch to w as the coefficient because in machine learning
#the algorthms modify the weights (w) to minimize the error
#Just a convention
```

```python
#let's look at an example with 3 features i.e.
#y = w1x1 + w2x2 + w3x3 + w0
```

```python
rng = np.random.RandomState(1)
```

```python
#Using rand generates numbers from the uniform distribution, [0,1]
X = 10 * rng.rand(1000,3)
X[:10]
```

```python
#Let's define the slopes
w1 = 1.5
w2 = -2.0
w3 = 1.0
```

```python
#Introduce the dot product method
#using the dot product to multiple the slopes i.e. the scalar
#values
np.dot(X[:5],[w1,w2,w3])
```

```python
XX = X[:,0]*w1 + X[:,1]*w2 + X[:,2]*w3 
```

```python
XX[:5]
```

```python
y = w1*X[:,0] + w2*X[:,1] + w3*X[:,2] + 0.5
```

```python
y[:5]
```

```python
model.fit(X,y)
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
#get the coefficient of determination of the prediction. 
#X is your test data and y is the true value
#Best possible score is 1.0
#https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#
score = model.score(X[900:],y[900:])
score
```
